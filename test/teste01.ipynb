```python
#testando exibição do relatório de veiculos cadastrados e estruturando a classe veiculo

import pandas as pd

dicionario_veiculos = {}

class Veiculos:
    def __init__(self, placa, marca, modelo, atual):
        self.placa = placa
        self.marca = marca
        self.modelo = modelo
        self.atual = atual

def cadastro_veiculo():
    placa = input("Digite a placa do veículo: ")
    marca = input("Digite a marca do veículo: ")
    modelo = input("Digite o modelo do veículo: ")
    atual = input("Digite a data de inscrição do veículo: ")

    dados = Veiculos(placa, marca, modelo, atual)
    dicionario_veiculos[placa] = {
        "marca": dados.marca,
        "modelo": dados.modelo,
        "atual": dados.atual
    }



cadastro_veiculo()

df_veiculos = pd.DataFrame.from_dict(dicionario_veiculos, orient='index')
display(df_veiculos)
```

```python
#criando protótipo da função que sera destinada a editar os valores dentro de uma tag
dicionario = {
    "001": "Corolla",
    "002": "Uno Escada",
    "003": "Astra"
}

def editar():
    print(6*"="+" Editar Veiculo "+6*"=")

    # 🔹 Loop até encontrar uma tag válida
    while True:
        tag = input("Digite a tag que deseja editar: ").strip()
        veiculo = dicionario.get(tag)

        if not veiculo:
            print(f"Tag '{tag}' não encontrada. Tente novamente.")
        else:
            break  # sai do loop quando achar uma tag válida

    # 🔹 Loop de edição
    while True:
        print("\nDigite (0) para sair")
        print("Digite (1) para editar o modelo do veículo")

        try:
            opcao = int(input("Digite a opção: "))
        except ValueError:
            print("Digite apenas números!")
            continue

        if opcao == 0:
            break

        elif opcao == 1:
            novo_modelo = input("Digite o novo modelo do veículo: ")
            dicionario[tag] = novo_modelo
            print(f"Modelo atualizado com sucesso!\n{dicionario}")

        else:
            print("Opção inválida!")

editar()
```

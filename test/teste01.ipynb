```python
import pandas as pd

dicionario_veiculos = {}

class Veiculos:
    def __init__(self, placa, marca, modelo, atual):
        self.placa = placa
        self.marca = marca
        self.modelo = modelo
        self.atual = atual

def cadastro_veiculo():
    placa = input("Digite a placa do veículo: ")
    marca = input("Digite a marca do veículo: ")
    modelo = input("Digite o modelo do veículo: ")
    atual = input("Digite a data de inscrição do veículo: ")

    dados = Veiculos(placa, marca, modelo, atual)
    dicionario_veiculos[placa] = {
        "marca": dados.marca,
        "modelo": dados.modelo,
        "atual": dados.atual
    }



cadastro_veiculo()

df_veiculos = pd.DataFrame.from_dict(dicionario_veiculos, orient='index')
display(df_veiculos)
```
